```python
# density function. Calculates absolute salinity and conservative temperature and from their density from TEOS-10

def get_density(salinity, temperature, pressure, longitude, latitude):
    
    SA = gsw.SA_from_SP(salinity, pressure, longitude, latitude)
    CT = gsw.CT_from_t(SA, temperature, pressure)
    
    return gsw.density.rho(SA, CT, pressure)
```

```python
density = get_density(woce_data.Salinity, woce_data.Temp, woce_data.Pres, woce_data.LON, woce_data.LAT)
density.isel(ZAX = 0).plot(cmap = 'inferno')
```

```python
# defining temperature mean and anomaly

temp = woce_data.Temp
salinity = woce_data.Salinity

temp_mean = temp.mean()
salinity_mean = salinity.mean()
density_mean = density.mean()

temp_anomly = temp - temp_mean
salinity_anomly = salinity - salinity_mean
density_anomly = density - density_mean
```
