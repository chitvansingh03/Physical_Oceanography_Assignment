```python
Tp_S_4 = xr.Dataset({'Tpoten' : Tpoten , 'Salinity' : Salinity})

Tp_val = Tp_S_4['Tpoten'].values.flatten()
S_val = Tp_S_4['Salinity'].values.flatten()

nbinsTp = 50
nbinsS = 50

Tp_val = Tp_val[S_val >= 30]
S_val = S_val[S_val >= 30]
Tp_val = Tp_val[~np.isnan(Tp_val) ]
S_val  = S_val[~np.isnan(S_val)]

hist, xedges, yedges = np.histogram2d(Tp_val, S_val, bins=[nbinsTp, nbinsS])

xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

dx = dy = 0.2 * np.ones_like(zpos)
dz = hist.ravel()

cmap = plt.get_cmap('viridis')  # You can choose any colormap you prefer

# Normalize the histogram values to range [0, 1] for colormap mapping
norm = plt.Normalize(hist.min(), hist.max())

# Create a color array using the colormap and normalized histogram values
colors = cmap(norm(hist.ravel()))

# Create the plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot 3D bars with colors based on the colormap
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average', color=colors)

# Add color bar for the colormap
mappable = plt.cm.ScalarMappable(norm=norm, cmap=cmap)
mappable.set_array([])
cbar = plt.colorbar(mappable, ax=ax, orientation='vertical')
cbar.set_label('Frequency')

# Set labels and title for the plot
ax.set_xlabel('Potential Temperature (C)')
ax.set_ylabel('Salinity (psu)')
ax.set_zlabel('Frequency')
ax.set_title('Potential Temperature-Salinity Volumetric Plot')


azimuth_angle = 50# Set the azimuth angle (in degrees)
ax.view_init(elev=30, azim=azimuth_angle)  # Adjust the azimuth angle
plt.savefig('Potential Temperature-Salinity Volumetric Plot.svg')
# Show the plot
plt.show()
```
